```python
#Si queremos que las imágenes sean mostradas en una ventana emergente quitar el inline
%matplotlib inline

# OpenCV-Python utiliza NumPy para el manejo de imágenes
import numpy as np
# cv2 es el módulo python para acceder a OpenCV 
import cv2 as cv
# Usamos las poderosas herramientas de graficación de matplotlib para mostrar imágenes, perfiles, histogramas, etc
import matplotlib.pyplot as plt
```

```python
#Cargo la imagen original y la imprimo
imgBGR = cv.imread('WP_B.png')
img = cv.cvtColor(imgBGR,cv.COLOR_BGR2RGB)
imgR,imgG,imgB = cv.split(img)
plt.figure(1)
plt.imshow(img)
plt.title('Original')
plt.show

#Realizo el algoritmo white patch
normR = 255/imgR.max()
normG = 255/imgG.max()
normB = 255/imgB.max()
auxR = np.uint8(imgR*normR)
auxG = np.uint8(imgG*normG)
auxB = np.uint8(imgB*normB)
imgRGB = cv.merge((auxR,auxG,auxB))
plt.figure(2)
plt.imshow(imgRGB)
plt.title('White Patch')
plt.show
```
