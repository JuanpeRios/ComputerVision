```python
#Si queremos que las imágenes sean mostradas en una ventana emergente quitar el inline
%matplotlib inline

# OpenCV-Python utiliza NumPy para el manejo de imágenes
import numpy as np
# cv2 es el módulo python para acceder a OpenCV 
import cv2 as cv
# Usamos las poderosas herramientas de graficación de matplotlib para mostrar imágenes, perfiles, histogramas, etc
import matplotlib.pyplot as plt
```

```python
def Cromaticas(img):
    imgR,imgG,imgB = cv.split(img)
    sumatoria = cv.add(imgR,cv.add(imgG,imgB))
    mask = sumatoria==0
    sumatoria = cv.add(sumatoria,1)
    auxR = imgR/sumatoria
    auxG = imgG/sumatoria
    auxB = imgB/sumatoria
    crom = np.stack([auxR,auxG,auxB],axis=2)
    return crom
    
```

```python
# img = cv.imread('mib-alien.jpg',cv.IMREAD_GRAYSCALE)
img1 = cv.imread('CoordCrom_1.png')
img = cv.cvtColor(img1,cv.COLOR_BGR2RGB)

# Imagen original
plt.figure(1)
plt.imshow(img)
plt.show
```

```python
crom = Cromaticas(img)
plt.figure(2)
plt.imshow(crom)
plt.show
```
